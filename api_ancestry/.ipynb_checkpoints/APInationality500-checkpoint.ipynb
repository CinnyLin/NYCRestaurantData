```python
TrowBus1913['First'] = TrowBus1913['Name'].str.split().str[1]
TrowBus1913['Last'] = TrowBus1913['Name'].str.split().str[0]
TrowBus1913.head(1)
```

```python
country_df = []
for i in range(0,499):
    first = TrowBus1913['First'][i]
    last = TrowBus1913['Last'][i]
    result = !curl -X GET "https://v2.namsor.com/NamSorAPIv2/api2/json/origin/{first}/{last}" -H "accept: application/json" -H "X-API-KEY: 5fe4dfb2938c5ebdaec25dc955df5da3"
    country_df.append(result[-1])
```

```python
country_df
```

```python
import numpy as np

country1 = []
country2 = []

for each in country_df:
    try:
        gen = each.split(',')[3].split(':')[1].strip('"')
        prob = each.split(',')[4].split(':')[1].strip('"')
    except:
        gen = 'error'
    
    country1.append(gen)
    country2.append(prob)
```

Replace data values in python.

[pandas.DataFrame.replace](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.replace.html)

`DataFrame.replace(self, to_replace=None, value=None, inplace=False, limit=None, regex=False, method='pad')[source]`

Parameters: to_replace: str, regex, list, dict, Series, int, float, or None

**dict**

For a DataFrame nested dictionaries, e.g., `{'a': {'b': np.nan}}`, are read as follows: look in column `‘a’` for the value `‘b’` and replace it with `NaN`. The value parameter should be None to use a nested dict in this way. You can nest regular expressions as well. Note that column names (the top-level dictionary keys in a nested dictionary) cannot be regular expressions.
