```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/census_1910_manhattan.csv')
#df.head()
```

```python
fig, ax = plt.subplots(4, 4, figsize=(20,20))

places = list(dict(df['Birthplace'].value_counts()).keys())

for k, place in enumerate(places[1:17]):
    x = df[df['Birthplace']==place]['Lon']
    y = df[df['Birthplace']==place]['Lat']
    if k<4: #k=0-3
        ax[0,k-1].scatter(x, y, label=place, alpha=0.01)
        ax[0,k-1].legend()
    elif k<8: #k=4-7
        ax[1,k-5].scatter(x, y, label=place, alpha=0.01)
        ax[1,k-5].legend()
    elif k<12: #k=8-11
        ax[2,k-9].scatter(x, y, label=place, alpha=0.01)
        ax[2,k-9].legend()
    else: #k=12-15
        ax[3,k-13].scatter(x, y, label=place, alpha=0.01)
        ax[3,k-13].legend()

plt.savefig('plots/top_ethnic_clusters_1910.png')
```

```python
df2 = pd.read_csv('datasets/1898_address_geocoded_cleaned.csv')
df2.head()
```

```python
df2['region_cleaned'].value_counts()
```

```python
#dict(df2['final'].value_counts())
```

```python
x1 = df[df['Birthplace']=='Germany']['Lon']
y1 = df[df['Birthplace']=='Germany']['Lat']

x2 = df2[df2['region_cleaned']=='Germanic']['Lon']
y2 = df2[df2['region_cleaned']=='Germanic']['Lat']

plt.figure(figsize=(10,10))
plt.scatter(x1, y1, label='population', alpha=0.01, marker='o')
plt.scatter(x2, y2, label='restaurant1898', alpha=0.3, marker='x', color='r')
plt.title('Germanic Restaurants in 1898')
plt.legend()

plt.savefig('plots/German_census_restaurant.png')
```
